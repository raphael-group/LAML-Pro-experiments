```python
def hamming_distance_missing(x, y, missing_char):
    hd = 0
    for i, xval in enumerate(x):
        yval = y[i]
        if xval != yval:
            if xval == missing_char and yval == missing_char:
                hd += 0
            elif xval == missing_char and yval != missing_char:
                hd += 0
            elif xval != missing_char and yval == missing_char:
                hd += 0
            else:
                hd += 1
    return hd/(len(x))
```

```python
# Function to calculate Hamming distances from pos2_cell15
def calculate_hamming_distances(df, reference_row='pos2_cell15'):
   
    # Extract the reference row as a numpy array
    reference_values = df.loc['pos2_cell15']
    
    # Calculate Hamming distance for each row compared to the reference row
    hamming_distances = []
    for index, row in df.iterrows():
        print(index)
        if index != reference_row:
            all_hd = []
            for i, refval in enumerate(reference_values):
                cmpval = row.values[i]
                hd = hamming_distance_missing([refval], [cmpval], -1)
                all_hd.append(hd)
            # distance = scipy.spatial.distance.hamming(reference_values, row.values)
            hamming_distances.append((row.name, all_hd))
            
    return hamming_distances
```

```python
colony = 2

# all_genotype_probs = []
baseMem_total_hd, laml2_total_hd = [], []

data_prob_dict = dict()
data_dict = dict()

for cell_name in overlapping_indices:    
    
    # cell_name = "pos2_cell27"
    # extract the genotype posterior probability (baseMem)
    genotype_prob = colony_dict[colony][cell_name]

    baseMem_cell_genotype = baseMem_subset.loc[cell_name].values
    laml2_cell_genotype = laml2_subset.loc[cell_name].values

    # get the non-missing positions
    keep_indices = np.where(genotype_prob != 0)
    baseMem_cell_genotype = baseMem_cell_genotype[keep_indices]
    laml2_cell_genotype = laml2_cell_genotype[keep_indices]
    genotype_prob = genotype_prob[keep_indices]

    imputed_match = []
    # compare laml2_imputed genotype to original genotype
    for i in range(len(genotype_prob)):
        match = 1 - hamming_distance_missing([baseMem_cell_genotype[i]], [laml2_cell_genotype[i]], -1)
        imputed_match.append(match)
    # 1 means match; 0 means mismatch

    genotype_prob = np.array(genotype_prob)
    match_indices = np.where(np.array(imputed_match) == 1)[0]
    diff_indices = np.where(np.array(imputed_match) != 1)[0]

    cell_avg_genotype_prob = np.mean(genotype_prob)
    # Extract genotype probabilities for matches and differences
    match_probs = genotype_prob[match_indices]
    diff_probs = genotype_prob[diff_indices]
    
    data_prob_dict[cell_name] = cell_avg_genotype_prob
    data_dict[cell_name] = [match_probs, diff_probs]
```

### weighted hamming distance

Goal: illustrate that the placement for laml2 is better. Do the weighted hamming distance. 

```python
df
```

```python
cols = list(np.array([f'baseMem_svm_state_{s}' for s in range(4)]))
cols
```

```python
# Reshape the dataframe values into a matrix (7 sites x 5 states)
# matrix_values = df.values.reshape(7, 5)
```

```python
posterior_prob_dict = dict()
for i, colony in df.groupby('position_number'):
    # print("position", i)
    i = int(i)
    posterior_prob_dict[i] = dict()
    
    for cell_id, cell in colony.groupby('cell_number'):
        posterior_prob_dict[i][int(cell_id)] = dict()
        for k, cassette in cell.groupby('integration_barcode'):
            
            for j, site in cassette.groupby('site_index'):
                probs = site[cols].values.flatten() # (1, 4)
                posterior_prob_dict[i][int(cell_id)][int(k*6 + j)] = probs
# dictionary [pos] [cell_idx] [site_index] = list of 4 state probabilities
```

```python
posterior_prob_dict
```

For every genotype:
    
    - find the sibling in each tree
    - compute the Pr(match) with the sibling in each tree
    - compute the Pr(different) with the sibling in each tree

```python
posterior_prob_dict[2][15][7]
```

```python
def cell_name_to_idx(name):
    # print(name)
    return name.split('_')[1][4:]

def compute_prob(posterior_prob_dict, pos, cell1, cell2, geno1, geno2):
    cell1 = int(cell_name_to_idx(cell1))
    cell2 = int(cell_name_to_idx(cell2))
    posterior_prob_dict = posterior_prob_dict[pos]
    pr_matches = []
    pr_mismatches = []
    
    geno1 = np.array(geno1)
    geno2 = np.array(geno2)
    
    whd = []
    # find the number of cassettes for which they are both non-missing
    shared_sites = set(posterior_prob_dict[cell1].keys()).intersection(set(posterior_prob_dict[cell2].keys()))
    shared_sites = list(shared_sites.intersection(range(len(geno1))))
    # print(len(shared_sites),  len(geno1), len(geno2))
    
    #print(len(shared_sites), len(geno1), len(geno2))
    for site_idx in shared_sites:
        # print(site_idx, type(site_idx))
        cell1_state_probdict = posterior_prob_dict[cell1][site_idx]
        cell2_state_probdict = posterior_prob_dict[cell2][site_idx]
        
        pr_match, pr_mismatch = 0, 0
        for state_idx in range(4):
#             print("cell1_state_probdict[state_idx]", cell1_state_probdict[state_idx])
#             print("cell2_state_probdict[state_idx]", cell2_state_probdict[state_idx])
#             print("cell1_state_probdict[state_idx]", cell1_state_probdict[state_idx].shape)
            
            pr_match += cell1_state_probdict[state_idx] * cell2_state_probdict[state_idx]
            for state2_idx in range(4): # order matters
                if state2_idx != state_idx:
                    pr_mismatch += cell1_state_probdict[state_idx] * cell2_state_probdict[state2_idx]
        pr_matches.append(pr_match)
        pr_mismatches.append(pr_mismatch)
        
        
        #if site_idx in geno1:
            
        state1 = geno1[site_idx]
        state2 = geno2[site_idx]

        if state1 == state2:
            whd.append(pr_match)
        else:
            whd.append(pr_mismatch)
            
    return np.mean(np.array(pr_matches)), np.mean(np.array(pr_mismatches)), np.mean(np.array(whd))
```
