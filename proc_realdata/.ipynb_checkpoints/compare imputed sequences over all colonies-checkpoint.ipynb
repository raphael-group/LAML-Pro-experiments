```python
import treeswift
import pandas as pd
import scipy
import ast
from Bio import Phylo
import skbio
import numpy as np
```

```python
def hamming_distance_missing(x, y, missing_char):
    hd = 0
    for i, xval in enumerate(x):
        yval = y[i]
        if xval != yval:
            if xval == missing_char and yval == missing_char:
                hd += 0
            elif xval == missing_char and yval != missing_char:
                hd += 0
            elif xval != missing_char and yval == missing_char:
                hd += 0
            else:
                hd += 1
    return hd/(len(x))
```

```python
msa_path = '/Users/gc3045/laml2_experiments/proc_realdata/baseMemoir/baseMemoir.msa.txt'
df = pd.read_csv(msa_path)
df['position'] = df['cell_names'].str.extract(r'^(pos\d+)_')
cols_of_interest = df.columns[1:-1]
```

```python
df
```

```python
# pos="2"

laml2_total_hd, baseMem_total_hd = 0, 0

pearson_results = {'baseMemoir': [], 'LAML2': []}
spearman_results = {'baseMemoir': [], 'LAML2': []}

for pos in ["2", "3", "4", "5", "6", "78", "910"]:
    ancestral_sequence_file = f"/Users/gc3045/laml2_experiments/proc_realdata/baseMemoir/impute_sequences/p{pos}_imputed_sequences.txt"
    laml2_tree_file = f"/Users/gc3045/laml2_experiments/proc_realdata/baseMemoir/baseMemoir_runs/LAML2_p{pos}_baseMem_trees.nwk"
    baseMem_tree_file = f"/Users/gc3045/laml2_experiments/proc_realdata/baseMemoir_p{pos}.nwk"

    baseMem_t = treeswift.read_tree_newick(baseMem_tree_file)
    for n in baseMem_t.traverse_leaves():
        n.label = f"pos{pos}_cell{n.label}"
    baseMem_distmat = baseMem_t.distance_matrix(leaf_labels=True)

    laml2_t = treeswift.read_tree_newick(laml2_tree_file)
    laml2_distmat = laml2_t.distance_matrix(leaf_labels=True)

    if pos == "78":
        df_pos = df[(df['position'] == "pos7") | (df['position'] == "pos8")]            
    elif pos == "910":
        df_pos = df[(df['position'] == "pos9") | (df['position'] == "pos10")]            
    else:
        df_pos = df[df['position'] == f"pos{pos}"]

    # extract all baseMem sequences
    baseMem_genotypes = dict()
    for _, row in df_pos.iterrows():
        cell_name = row['cell_names']
        if pos == "78":
            cell_name = "pos78_" + cell_name.split('_')[1]
        elif pos == "910":
            cell_name = "pos910_" + cell_name.split('_')[1]
        genotype = row[cols_of_interest].values.tolist()
        baseMem_genotypes[cell_name] = genotype
    
    baseMem_genotype_distmat = dict()
    for cell_name_i in baseMem_genotypes:
        for cell_name_j in baseMem_genotypes:
            if cell_name_i not in baseMem_genotype_distmat:
                baseMem_genotype_distmat[cell_name_i] = dict()
            genotype_1 = baseMem_genotypes[cell_name_i]
            genotype_2 = baseMem_genotypes[cell_name_j]
            baseMem_genotype_distmat[cell_name_i][cell_name_j] = hamming_distance_missing(genotype_1, genotype_2, -1)

    laml2_genotypes = dict()
    with open(ancestral_sequence_file, "r") as f:
        lines = f.readlines()
        for line in lines:
            cell_name = line.split()[0]
            rest = line[len(cell_name)+1:]
            imputed_genotype = ast.literal_eval(rest)
            imputed_genotype = [x[0] for x in imputed_genotype]
            laml2_genotypes[cell_name] = imputed_genotype
    laml2_genotype_df = pd.DataFrame(laml2_genotypes).T

    laml2_genotype_distmat = dict()
    for cell_name_i in laml2_genotypes:
        for cell_name_j in laml2_genotypes:
            if cell_name_i not in laml2_genotype_distmat:
                laml2_genotype_distmat[cell_name_i] = dict()
            genotype_1 = laml2_genotypes[cell_name_i]
            genotype_2 = laml2_genotypes[cell_name_j]
            laml2_genotype_distmat[cell_name_i][cell_name_j] = hamming_distance_missing(genotype_1, genotype_2, -1)

    # format baseMem tree distance matrix into dataframe
    baseMem_distmat_df = pd.DataFrame.from_dict(baseMem_distmat, orient='index').fillna(0)
    # format baseMem genotype distance matrix into dataframe
    baseMem_genotype_distmat_df = pd.DataFrame.from_dict(baseMem_genotype_distmat, orient='index')
    # filter to match baseMem_distmat_df
    common_labels = list(set(baseMem_distmat_df.index) & set(baseMem_genotype_distmat_df.index))
    baseMem_distmat_df = baseMem_distmat_df.loc[common_labels, common_labels]
    baseMem_genotype_distmat_df = baseMem_genotype_distmat_df.loc[common_labels, common_labels]
    ordered_labels = sorted(baseMem_distmat_df.columns)
    baseMem_genotype_distmat_df = baseMem_distmat_df.reindex(index=ordered_labels, columns=ordered_labels)
    baseMem_genotype_distmat_df = baseMem_genotype_distmat_df.reindex(index=ordered_labels, columns=ordered_labels)

    # format laml2 genotype distance matrix into dataframe
    laml2_genotype_distmat_df = pd.DataFrame.from_dict(laml2_genotype_distmat, orient='index')
    ordered_labels = sorted(laml2_genotype_distmat_df.columns)
    laml2_genotype_distmat_df = laml2_genotype_distmat_df.reindex(index=ordered_labels, columns=ordered_labels)
    # format laml2 tree distance matrix into dataframe
    laml2_distmat_df = pd.DataFrame.from_dict(laml2_distmat, orient='index').fillna(0)
    ordered_labels = sorted(laml2_distmat_df.columns)
    laml2_distmat_df = laml2_distmat_df.reindex(index=ordered_labels, columns=ordered_labels)

    # pearson
    # print("pearson")
    pearson_results['baseMemoir'].append(skbio.stats.distance.mantel(baseMem_distmat_df, baseMem_genotype_distmat_df, method="pearson", permutations=10000)[:2])
    pearson_results['LAML2'].append(skbio.stats.distance.mantel(laml2_distmat_df, laml2_genotype_distmat_df, method="pearson", permutations=10000)[:2])
    
    # spearman
    # print("spearman")
    spearman_results['baseMemoir'].append(skbio.stats.distance.mantel(baseMem_distmat_df, baseMem_genotype_distmat_df, method="spearman", permutations=10000)[:2])
    spearman_results['LAML2'].append(skbio.stats.distance.mantel(laml2_distmat_df, laml2_genotype_distmat_df, method="spearman", permutations=10000)[:2])

    # print(sum(laml2_genotype_distmat_df.sum()))
    laml2_total_hd += sum(laml2_genotype_distmat_df.sum())
    baseMem_total_hd += sum(baseMem_distmat_df.sum())
    break
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming pearson_results and spearman_results are already populated with the Mantel test results
# Example structure:
# pearson_results = {'baseMemoir': [(correlation1, pvalue1), (correlation2, pvalue2), ...], 'LAML2': [...]}
# spearman_results = {'baseMemoir': [(correlation1, pvalue1), (correlation2, pvalue2), ...], 'LAML2': [...]}

# Positions for x-axis labels
positions = ["P" + str(pos) for pos in ["2", "3", "4", "5", "6", "78", "910"]]

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Pearson plot
x = np.arange(len(positions))
width = 0.35

rects1 = ax1.bar(x - width / 2, [item[0] for item in pearson_results['baseMemoir']], width, label='baseMemoir', color='#20808D')
rects2 = ax1.bar(x + width / 2, [item[0] for item in pearson_results['LAML2']], width, label='LAML2', color='#A84B2F')

ax1.set_ylabel('Pearson Correlation Coefficient', fontname='Times New Roman', fontsize=14)
ax1.set_title('Pearson Corr. between Tree and Genotype Distances', fontname='Times New Roman', fontsize=16)
ax1.set_xlabel('Colonies', fontname='Times New Roman', fontsize=16)
ax1.set_xticks(x)
ax1.set_xticklabels(positions, fontname='Times New Roman', fontsize=14)
ax1.legend(prop={'family': 'Times New Roman', 'size': 12})

for yline in np.arange(-0.2, 1.0, 0.1):
    ax1.axhline(y=yline, color='gray', linestyle='--')

def autolabel(rects, pvalues, above=True):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect, pvalue in zip(rects, pvalues):
        height = rect.get_height()
        if above:
            ax1.annotate(f'p={pvalue:.4f}',
                     xy=(rect.get_x() + rect.get_width() / 2, height),
                     xytext=(0, 3),  # Offset text above the bar
                     textcoords="offset points", color='#A84B2F',
                     ha='center', va='bottom', fontname='Times New Roman', fontsize=12)
        else:
            ax1.annotate(f'p={pvalue:.4f}',
                     xy=(rect.get_x() + rect.get_width() / 2, -0.15),
                     #xytext=(0, -5),  # Offset text above the bar
                     textcoords="offset points", color='#20808D',
                     ha='center', va='bottom', fontname='Times New Roman', fontsize=12)

autolabel(rects1, [item[1] for item in pearson_results['baseMemoir']], above=False)
autolabel(rects2, [item[1] for item in pearson_results['LAML2']])

# Spearman plot
rects3 = ax2.bar(x - width / 2, [item[0] for item in spearman_results['baseMemoir']], width, label='baseMemoir', color='#20808D')
rects4 = ax2.bar(x + width / 2, [item[0] for item in spearman_results['LAML2']], width, label='LAML2', color='#A84B2F')

ax2.set_ylabel('Spearman Correlation Coefficient', fontname='Times New Roman', fontsize=14)
ax2.set_title('Spearman Corr. between Tree and Genotype Distances', fontname='Times New Roman', fontsize=16)
ax2.set_xlabel('Colonies', fontname='Times New Roman', fontsize=16)
ax2.set_xticks(x)
ax2.set_xticklabels(positions, fontname='Times New Roman', fontsize=14)
ax2.legend(prop={'family': 'Times New Roman', 'size': 12})

for yline in np.arange(-0.2, 1.0, 0.1):
    ax2.axhline(y=yline, color='gray', linestyle='--')

def autolabel_spearman(rects, pvalues, above=True):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect, pvalue in zip(rects, pvalues):
        height = rect.get_height()
        if above:
            ax2.annotate(f'p={pvalue:.4f}',
                     xy=(rect.get_x() + rect.get_width() / 2, height),
                     xytext=(0, 3),  # Offset text above the bar
                     textcoords="offset points", color='#A84B2F',
                     ha='center', va='bottom', fontname='Times New Roman', fontsize=12)
        else:
            ax2.annotate(f'p={pvalue:.4f}',
                     xy=(rect.get_x() + rect.get_width() / 2, -0.15),
                     #xytext=(0, -5),  # Offset text above the bar
                     textcoords="offset points", color='#20808D',
                     ha='center', va='bottom', fontname='Times New Roman', fontsize=12)
autolabel_spearman(rects3, [item[1] for item in spearman_results['baseMemoir']], above=False)
autolabel_spearman(rects4, [item[1] for item in spearman_results['LAML2']])

fig.tight_layout()
plt.savefig("/Users/gc3045/laml2_experiments/figures/distance_correlation_all_positions.pdf")
```
