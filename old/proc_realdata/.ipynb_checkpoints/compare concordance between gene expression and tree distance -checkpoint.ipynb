```python

def compute_parsimony_cost(tree, leaf_label_df, states):

    cost_dict = dict()
    idx = 0
    for n in tree.traverse_postorder(): 
        print(n.label,f'node_{idx}' )
        n.label = f'node_{idx}'
        idx += 1
        
    for n in tree.traverse_postorder(): 
        if n.is_leaf():
            cost_dict[n.label] = dict()
            for state in states:
                idx = leaf_label_df.index[leaf_label_df['cell_idx'] == state].tolist()
                print(idx, state, n.label)
                obs_state = leaf_label_df.loc[idx, 'state_idx']
                if obs_state == state:
                    cost_dict[n][state] = 0
                else:
                    cost_dict[n][state] = 10000000
        else:
            u, v = n.children
            for state in states:
                cost_dict[n.label][state] = cost_dict[u.label][min(cost_dict[u.label], key=cost_dict[u.label].get)] + cost_dict[v.label][min(cost_dict[v.label], key=cost_dict[v.label].get)] 
        return cost_dict[tree.root.label][min(cost_dict[tree.root.label], key=cost_dict[tree.root.label].get)]
                    
                    
```

```python
laml2_folder_path = "/Users/gc3045/laml2_experiments/proc_realdata/baseMemoir/baseMemoir_runs"
folder_path = "/Users/gc3045/laml2_experiments/proc_realdata/baseMemoir"
tree_files = glob.glob(os.path.join(folder_path, "*starting_tree.nwk"))

states = [1, 2, 3, 4, 5]
tree_objs = []

for f in tree_files:
    pos = os.path.basename(f).split('_')[0]
    print('pos', pos)
    print('basemem tree_file', f)
    
    laml_tfile = f'{laml2_folder_path}/LAML2_{pos}_baseMem_trees.nwk'
    print('laml_tree_file', laml_tfile)
    
    baseMem_tree = treeswift.read_tree_newick(f)
    laml_tree = treeswift.read_tree_newick(laml_tfile)
    
    baseMem_distmat = baseMem_tree.distance_matrix(leaf_labels=True)
    laml2_distmat = laml_tree.distance_matrix(leaf_labels=True)

#     print(list([l.label for l in laml_tree.traverse_leaves()]))
    assert len(set([l.label for l in laml_tree.traverse_leaves()])) == len(set([l.label for l in baseMem_tree.traverse_leaves()]))
    
    # compute the cell state parsimony cost
    
    leaf_label_df = all_df[str(pos[1:])]
    compute_parsimony_cost(laml_tree, leaf_label_df, states)
    
```
